```python
import numpy as np
import pandas as pd

# Moduule for processig text data
import nltk
```

```python
bio = pd.read_csv('../data/bios.csv')
```

```python
bio.head()
```

```python
# Function to lemmatize text data and remove stopwords and punctuations

import re
from nltk.tokenize import word_tokenize # import tokeninzing function
from nltk.stem import WordNetLemmatizer # import lemmatizinf function
from nltk.corpus import stopwords # import stopwords
stwds = stopwords.words('english') # list of stopwords in english language
lemma = WordNetLemmatizer() # function to lemmatize words

def tokenize(text):
    text = re.sub('[^a-zA-Z]',' ',text) # to replace puncatuations and anything except A-z and a-z with ' ' (space)
    text = text.lower() # converting text to lower case
    text = word_tokenize(text) # tokening text data into list
    text = [lemma.lemmatize(word) for word in text if word not in stwds] # lemmatizing the words in list
    text = ' '.join(text) # joining the words in list
    return text
```

```python
# applyting tokenize function to bio column of dataframe

bio = bio.bio.apply(tokenize)
bio
```

```python
# Import CountVectorizer function to count number of time a words appears in a sentence

from sklearn.feature_extraction.text import CountVectorizer
vectorise = CountVectorizer()
```

```python
# Transofrming the text data into array using CountVectoriser function

X = vectorise.fit_transform(bio).toarray()
X
```

```python
# Converting data type of array to reduce size

X = X.astype('int16')
```

```python
X[0].dtype
```

```python
# Convertinng array into dataframe 

X = pd.DataFrame(X,columns=vectorise.get_feature_names_out())
X
```

```python
# Importing other preprocessed features

df = pd.read_csv('../data/features.csv')
df.head()
```

```python
# Concatenate the encoded features and vetorised bio 

df = pd.concat((df,X),axis=1)
```

```python
df.head()
```

```python
# Saving the all the preprocessed and vectorised features into single csv file

df.to_csv('../data/refined_data.csv',index=False)
```
