```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Bifurcating dataset into different datasets as per their use

# to concat it to match matrix
user_id = pd.DataFrame(df['user_id']) 

# vectorising bio seperately
bio = pd.DataFrame(df['bio']) 

# to perform one hot encoding
lang = pd.DataFrame(df['language']) 

# no need to preprocess these features
data = df[['sex','orientation','age','height','education_level']] 

# remaining feature to preprocess
df = df.drop(['user_id','username','age','height','sex','orientation','education_level','bio','language'],axis=1) 
```

```python
# Determine shape of df

df.shape
```

```python
# Lebel Encoding features of df

from sklearn.preprocessing import LabelEncoder
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
# Concating df to data's numerical features

df = pd.concat((data.drop(['sex','orientation'],axis=1),df),axis=1)
df.shape
```

```python
# Check total number of unique values in language feature

lan = []
for i in range(len(lang)):
    lan.extend(lang.language[i].split(', '))
lan = set(lan)
len(lan)
```

```python
# One hot encoding language feature
# split the value in each rw by ', ' and mark the value as one if j is present in the row

lang_dict = {}
for i in lan:
    l = []
    for j in lang.language:
        if i in j.split(', '):
            l.append(1)
        else:
            l.append(0)
    lang_dict[i]=l
lan = pd.DataFrame(lang_dict)

# Concate the lan (language) dataframe with df
df = pd.concat((df,lan),axis=1)
df.shape
```

```python
# Saving the preprocessed data as csv file
df.to_csv('../data/features.csv', index=False)
```

```python
# Saving features in csv files to use in future as per requirement to avoid memory allocation to unnecessary features
user_id.to_csv('../data/id.csv', index=False)
bio.to_csv('../data/bios.csv', index=False)
orientation.to_csv('../data/preference.csv', index=False)
```
